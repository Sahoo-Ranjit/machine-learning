```python
df_country = pd.read_excel("/kaggle/input/eda-zomatodataset/Dataset File/Country-Code.xlsx")
```

```python
df1 = pd.merge(df,df_country,on='Country Code')
```

```python
df1.tail()
```

```python
df1.dtypes
```

```python
country_names = df1.Country.value_counts().index
```

```python
country_names
```

```python
country_values = df1.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

**Observation**
* India has the most of the transaction happening followed by United States and United Kingdom 

```python
df2 = df1.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
df2
```

**Observation**
1. When the Rating is between 4.5-4.9---->Excelent
2. When the Rating is between  4.0-4.5--->very good
3. When the Rating is between  3.5-3.9---->good
4. When the Rating is between  3.0-3.5---->Average
5. When the Rating is between  1.8-2.4---->poor
6. Many people have not given the rating to the app thats why many ratings can be seen as 0

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=df2)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=df2,hue='Rating color',palette=['white','red','orange','yellow','green','blue'])
```

```python
sns.countplot(x='Rating color',data=df2,palette=['white','red','orange','yellow','green','blue'])
```
